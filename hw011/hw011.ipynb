```python
from sympy import *
from sympy.plotting import plot

# f(x) = 5x^2 + 10x - 30
```

#### Определить корни

```python
x = Symbol('x')
y = 5*x**2 + 10*x - 30
solve(y)
```

#### Найти интервалы, на которых функция возрастает

```python
d = y.diff(x)
solve(d > 0)
```

#### Найти интервалы, на которых функция убывает

```python
solve(d < 0)
```

#### Построить график

```python
plot(y)
y
```

#### Вычислить вершину

```python
ab = solve(d, x)[0]
[ab, y.subs(x, ab)]
```

#### Определить промежутки, на которых f > 0

```python
solve(y > 0)
```

#### Определить промежутки, на которых f < 0

```python
solve(y < 0)
```
